# Basic clustering

## importing libraries

```python
%matplotlib inline
import numpy as np
from sklearn.cluster import KMeans
from matplotlib import pyplot as plt
```

## creating artificial data set

```python
n_samples = 500

X = np.concatenate((
    np.random.normal((-2, -2), size=(n_samples, 2)),
    np.random.normal((2, 2), size=(n_samples, 2))
))

plt.figure(figsize=(5, 5))
plt.scatter(X[:,0], X[:,1], alpha=.5)
plt.show()
```

# fitting the k-means

```python
model = KMeans(n_clusters=2, max_iter=1)
model.fit(X)
```

# showing results

```python
# outputs
fcm_centers = model.cluster_centers_
fcm_labels = model.predict(X)

# plot result
f, axes = plt.subplots(1, 2, figsize=(11,5))
axes[0].scatter(X[:,0], X[:,1], alpha=.1)
axes[1].scatter(X[:,0], X[:,1], c=fcm_labels, alpha=.1)
axes[1].scatter(fcm_centers[:,0], fcm_centers[:,1], marker="+", s=500, c='k')
plt.savefig('images/basic-clustering-output.jpg')
plt.show()
```
